```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
n=20000
m= 20
p = 0.1
g = nx.powerlaw_cluster_graph(n, m, p, seed=None)
#nx.draw(g)
#plt.show()
print(g.size())
dgs = [d for k,d in g.degree()]
print(np.mean(dgs))
x=plt.hist(dgs)
x=plt.hist(dgs[1:])
key = 'n_'+str(n)+'_m_'+str(m)+'_p_'+str(p)+'_num_edges_'+str(g.size())
nx.write_edgelist(g,'power_law_'+key+'.edgelist',data=False)
```

```python
import numpy as np
dgs = np.array(list(sorted(dgs,reverse=True)))
print(dgs[:10])
x=plt.hist(dgs[dgs>1000])
```

```python
N = 100000
p = 10
p1 = 0.001
p2 = 0.0001
sizes = [int(N/p)]*p
P = np.array([[p2/(p-1)]*p]*p)
print(P)
for i in range(p):
    P[i][i] = p1
    print(sum(P[i]))
print(P)
#p = [[0.9,0.05,0.05],[0.05, 0.9, 0.05], [0.05,0.05,0.9]]
print(np.array(p))
g = nx.stochastic_block_model(sizes, P, nodelist=None, seed=0, directed=True, selfloops=False, sparse=True)
#nx.draw(g)
print(g.size())
#plt.plot()
#plt.show()
dgs = [d for k,d in g.degree()]
print(np.mean(dgs))
plt.hist(dgs)
key = 'n_'+str(N)+'_parts_'+str(p)+'_p1_'+str(p1) +'_p2_'+str(p2)+'_num_edges_'+str(g.size()) + '_'
print(key)
nx.write_edgelist(g,'sbm_'+key+'.edgelist',data=False)
```
